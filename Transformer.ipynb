`pipeline()` : connect a model with it necessary preprocessing and postpreprocessing model

```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis") # model is downloaded and cached while creating object of pipeline
classifier("I am waiting for my last breadth")
classifier("The life is moving so fast")

# 3 main step is involve
# 1 text preprocessed
# 2 preproceesed input pass to the model
# 3 post processed the model predictions
```
