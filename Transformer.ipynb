# zero-shot-learning

in this method we label the unlabelled text by giving the label list in this they automatically give scoring to the label

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification")
classifier("I am living life that is no one assume",
candidate_labels = ["happy","sad","unpredicted"])
```

```python
classifier(
    "stomach ulcer tablet is high dose and getting me in trouble",
    candidate_labels = ["neutral","good","bad","not good"]
)
```

## fill-mask

this is pipeline use for fill the dash in the sentence

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker(
    "I would like <mask> in tech",
    top_k=3  # display the possibility in the mask
)
```
