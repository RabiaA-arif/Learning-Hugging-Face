## Named entity recognition

ner tell us relationship of person ,location and organization

imagine reading in article and automatically detect and highlight all the people ,location and oraganization in which they work

### why ner is usefule:
`information extraction`
`searching and indexing`
`virtual chatbot`
`knowledege graph`

```python
from transformers import pipeline

ner_pipeline = pipeline("ner",aggregation_strategy="simple")  # grouped_entities=True is remove from hugging face library
# typeerror : because grouped_entities is deprecated from lbrary
text = "my name is Rabia ,i live in pakistan,i work on google."

ner_results = ner_pipeline(text)

for en in ner_results:
  print(f"word : {en['word']}")
  print(f"Entitiy group :{en['entity_group']}")
  print(f"Score : {en['score']}")
```

```python
ner = pipeline("ner",aggregation_strategy="simple")
ner("I live in pakistan and study in AI,i am rabia")
```

```python
import pandas as pd

# Rerun the NER pipeline to get the output
output = ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")

# Create a DataFrame from the output
df_output = pd.DataFrame(output)
display(df_output)
```

## method 2 of ner technique:more control way of ner

```python
from transformers import AutoTokenizer ,AutoModelForTokenClassification
import torch
```

```python
model_name = "OpenMed/privacy-filter-multilingual-v2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

text = "My name is DR Ayesha ,and i live in lahore i work at dow hospital"

inputs = tokenizer(text,return_tensors = "pt")

with torch.no_grad():
  outputs = model(**inputs)

predictions = torch.argmax(outputs.logits,dim=2)

print("Processing token level")
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
predicted_labels = [model.config.id2label[prediction.item()] for prediction in predictions[0]]

for token , label in zip(tokens,predicted_labels):
  print(f"{token} : {label}")

```
